```python
# write a program to export the result of elevation surface down thickness (50 feet interval * 10 intervals)
def Elevation_surfacedown(df, interval=50, layers = 1):
    #records1 = intervals that go through the top of the layer and bottom is within layer
    records1 = df.loc[(df['TopEle'] > df['SurEle']-interval*(layers-1)) & (df['BotEle'] < df['SurEle']-interval*(layers-1)) & (df['BotEle'] >= df['SurEle']-interval*layers) & (df['BotEle'] <= df['TopEle'])]
    thickness1 = pd.DataFrame((records1['SurEle']-interval*(layers-1)-records1.loc[: , 'BotEle']) * records1['Target'])
    thickness1.columns = ['thickness']
    records1 = records1.join(thickness1)
    
    #records2 = entire interval is within layer
    records2 = df.loc[(df['TopEle'] <= df['SurEle']-interval*(layers-1)) & (df['BotEle'] >= df['SurEle']-interval*layers) & (df['BotEle'] <= df['TopEle'])]
    thickness2 = pd.DataFrame((records2.loc[: , 'TopEle'] - records2.loc[: , 'BotEle']) * records2['Target'])
    thickness2.columns = ['thickness']
    records2 = records2.join(thickness2)
    
    #records3 = intervals with top within layer and bottom of interval going through bottom of layer
    records3 = df.loc[(df['TopEle'] > df['SurEle']-interval*layers) & (df['BotEle'] < df['SurEle']-interval*layers) & (df['TopEle'] <= df['SurEle']-interval*(layers-1)) & (df['BotEle'] <= df['TopEle'])]
    thickness3 = pd.DataFrame((records3.loc[: , 'TopEle'] - (records3['SurEle']-interval*layers))*records3['Target'])
    thickness3.columns = ['thickness']
    records3 = records3.join(thickness3)
    
    #records4 = interval goes through entire layer
    records4 = df.loc[(df['TopEle'] > df['SurEle']-interval*(layers-1)) & (df['BotEle'] < df['SurEle']-interval*layers) & (df['BotEle'] <= df['TopEle'])] 
    thickness4 = pd.DataFrame((records4.loc[: , 'BotEle']*0 + interval) * records4['Target'])
    thickness4.columns = ['thickness']
    records4 = records4.join(thickness4)
    
    res = records1.append(records2).append(records3).append(records4)
    res_df = res.groupby(['API_Number','Lat','Lon']).sum()[['thickness']]
    thickness_per = pd.DataFrame(res_df.loc[: , 'thickness']/interval)
    thickness_per.columns = ['thick_per']
    res_df = res_df.join(thickness_per)
    
    layer = pd.DataFrame(res_df.loc[: , 'thickness']*0+layers)
    layer.columns = ['layer']
    res_df = res_df.join(layer)
    
    return res, res_df
```

```python
# experiment with the first layer
res, res_df = Elevation_surfacedown(df, layers=1)
res
```

```python
# export the results from the range we want to csv
for i in np.arange(1,12):
    res, res_df = Elevation_surfacedown(df, layers = i)
    outputname = 'Elevation_Surface_Table\\tables\\SurfaceLayer_'+str(i)+'.csv'
    res_df.to_csv(outputname) 
    # export df as table
```

```python
# export the whole record (x,y,z)
res, res_df_whole = Elevation_surfacedown(df, layers = 1)
for i in np.arange(2,12):
    res, res_df = Elevation_surfacedown(df, layers = i)
    res_df_whole = res_df_whole.append(res_df)
outputname = 'Elevation_Surface_Table\\Surface_whole.csv'
res_df_whole.to_csv(outputname)
```

```python
# write a program to export the result of elevation thickness (25 feet interval)
def Elevation_thickness(df, bot_height, interval=25):
    records1 = df.loc[(df['TopEle'] > bot_height+interval) & (df['BotEle'] < bot_height+interval) & (df['BotEle'] >= bot_height) & (df['BotEle'] <= df['TopEle']) & (df['Target'].isin([1]))]
    thickness1 = bot_height+interval - pd.DataFrame(records1.loc[: , 'BotEle'])
    thickness1.columns = ['thickness']
    records1 = records1.join(thickness1)
    
    records2 = df.loc[(df['TopEle'] <= bot_height+interval) & (df['BotEle'] >= bot_height) & (df['BotEle'] <= df['TopEle']) & (df['Target'].isin([1]))]
    thickness2 = pd.DataFrame(records2.loc[: , 'TopEle'] - records2.loc[: , 'BotEle'])
    thickness2.columns = ['thickness']
    records2 = records2.join(thickness2)
    
    records3 = df.loc[(df['TopEle'] > bot_height) & (df['BotEle'] < bot_height) & (df['TopEle'] <= bot_height+interval) & (df['BotEle'] <= df['TopEle']) & (df['Target'].isin([1]))]
    thickness3 = pd.DataFrame(records3.loc[: , 'TopEle']) - bot_height
    thickness3.columns = ['thickness']
    records3 = records3.join(thickness3)
    
    records4 = df.loc[(df['TopEle'] > bot_height+interval) & (df['BotEle'] < bot_height) & (df['BotEle'] <= df['TopEle']) & (df['Target'].isin([1]))] 
    thickness4 = pd.DataFrame(records4.loc[: , 'BotEle']*0) + interval
    thickness4.columns = ['thickness']
    records4 = records4.join(thickness4)
    
    res = records1.append(records2).append(records3).append(records4)
    res_df = res.groupby(['API_Number','Lat','Lon']).sum()[['thickness']]
    thickness_per = pd.DataFrame(res_df.loc[: , 'thickness']/interval)
    thickness_per.columns = ['thick_per']
    res_df = res_df.join(thickness_per)
    
    bot_height = pd.DataFrame(res_df.loc[: , 'thickness']*0+bot_height)
    bot_height.columns = ['bot_height']
    res_df = res_df.join(bot_height)

    top_height = pd.DataFrame(res_df.loc[: , 'bot_height']+interval)
    top_height.columns = ['top_height']
    res_df = res_df.join(top_height)
    
    return res, res_df
```

```python
# experiment with 100 feet elevation
bot_height = 200
res, res_df = Elevation_thickness(df, bot_height)
res
```

```python
# export the results from the range we want to csv
for i in np.arange(200,1000, 25):
    res, res_df = Elevation_thickness(df, i)
    outputname = 'Elevation_Thick_Table\\tables\\Thickness_'+str(i)+'.csv'
    res_df.to_csv(outputname) 
    # export df as table
```

```python
# export the whole record (x,y,z)
res, res_df_whole = Elevation_thickness(df, 200)
for i in np.arange(225,1000, 25):
    res, res_df = Elevation_thickness(df, i)
    res_df_whole = res_df_whole.append(res_df)
outputname = 'Elevation_Thick_Table\\Thickness_whole.csv'
res_df_whole.to_csv(outputname) 
```
